Reading the table from the wikipedia page

```python
req = requests.get("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")

soup = bs(req.content,'lxml')

table = soup.find_all('table')[0]

df = pd.read_html(str(table))

neighborhood=pd.DataFrame(df[0])

neighborhood
```

## Removing the not assigned values

```python
df = neighborhood
df["Borough"].replace("Not assigned",np.nan,inplace=True)
df.dropna(subset=["Borough"],axis=0,inplace=True)
df.reset_index(inplace=True)
df
```

## Final DataFrame

```python
df=df[["Postal Code","Borough","Neighborhood"]]
df
```

```python
import geocoder # import geocoder

# initialize your variable to None
lat_lng_coords = None
postal_code = df["Postal Code"]
print(postal_code)
i=0
longitude = list()
latitude = list()
# loop until you get the coordinates
while(lat_lng_coords is None):
    g = geocoder.google('{}, Toronto, Ontario'.format(postal_code[i]))
    lat_lng_coords = g.latlng
    latitude.append(str(lat_lng_coords[0]))
    longitude.append(str(lat_lng_coords[1]))
    i+=1
print(longitude)
```
